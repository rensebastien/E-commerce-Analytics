```python
import zipfile
import gdown
import pandas as pd
import matplotlib.pyplot as plt
```

```python
 #Downloading Images and Annotations

zip_file_id = '1haRZvnDvoPEJOI9ZYaTtkBOixoBatikH'
gdown.download(f'https://drive.google.com/uc?id={zip_file_id}', 'ecommerce.zip', quiet=False)

with zipfile.ZipFile('ecommerce.zip', 'r') as zip_ref:
    zip_ref.extractall('data')
```

```python
customers=pd.read_csv('data/olist_customers_dataset.csv')
order_items=pd.read_csv('data/olist_order_items_dataset.csv')
locations=pd.read_csv('data/olist_geolocation_dataset.csv')
payments=pd.read_csv('data/olist_order_payments_dataset.csv')
review_ratings=pd.read_csv('data/olist_order_reviews_dataset.csv')
orders=pd.read_csv('data/olist_orders_dataset.csv')
products=pd.read_csv('data/olist_products_dataset.csv')
sellers=pd.read_csv('data/olist_sellers_dataset.csv')
```

```python
print('customers dataset  ')
display(customers.head())

print('\n order items dataset ')
display(order_items.head())

print('\n locations dataset ')
display(locations.head())

print('\n payments dataset ')
display(payments.head())

print('\n review_ratings dataset ')
display(review_ratings.head())

print('\n orders dataset')
display(orders.head())

print('\n products dataset')
display(products.head())

print('\n sellers dataset')
display(sellers.head())
```

```python
# Calcul du nombre de commandes par mois
orders["order_purchase_timestamp"] = pd.to_datetime(orders["order_purchase_timestamp"])
orders["year_month"] = orders["order_purchase_timestamp"].dt.to_period("M")

orders.groupby("year_month")["order_id"].count().plot(figsize=(12,6))
plt.title("Nombre de commandes par mois")
plt.xlabel("Date")
plt.ylabel("Commandes")
plt.show()
```

```python
# Fusion des avis avec les items et produits
merged = review_ratings.merge(orders, on="order_id").merge(order_items, on="order_id").merge(products, on="product_id")
category_ratings = merged.groupby("product_category_name")["review_score"].mean().sort_values(ascending=False).head(10)

category_ratings.plot(kind="bar", figsize=(12,6), color="blue")
plt.title("Top 10 catégories avec la meilleure note moyenne")
plt.xlabel("Catégorie de produit")
plt.ylabel("Note moyenne")
plt.ylim(3.5, 5)      # la note moyenne va rarement en dessous de 3
plt.yticks([3.5, 3.7, 3.9, 4.1, 4.3, 4.5, 4.7, 4.9,5])
plt.xticks(rotation=45, ha='right')
plt.show()


```

```python
# Construction d'un histogramme des 20 villes où il y a le plus de clients différents

# Nombre de clients par ville
clients_par_ville = customers.groupby("customer_city")["customer_id"].nunique().sort_values(ascending=False)
print(clients_par_ville.head(20))


clients_par_ville.head(20).plot(kind="bar", figsize=(12,6))
plt.title("Nombre de clients par ville")
plt.xlabel("Ville")
plt.ylabel("Nombre de clients différents")
plt.yticks(rotation=45,ha='right')
plt.xticks(rotation=45, ha='right')
plt.show()
```

L'analyse de cet histogramme met en lumière une concentration de clients dans une ville dominante qui est Sao Paolo. Rio de Janeiro suit dans le classement mais il y a tout de même une forte disparité entre Sao Paolo et Rio.


```python
# Construction d'un graphique à barres qui montre suivant l'année le total des ventes pour les 3 catégories de produits les plus vendus

# transformation en date
orders["order_purchase_timestamp"] = pd.to_datetime(orders["order_purchase_timestamp"])

# récupération des années
orders["year"] = orders["order_purchase_timestamp"].dt.year

merged = (
    order_items.merge(products[["product_id","product_category_name"]], on="product_id", how="inner")
    .merge(orders[["order_id", "order_purchase_timestamp", "year"]], on="order_id", how="inner")
)

ventes_par_produit = merged.groupby("product_category_name")["price"].sum().sort_values(ascending=False)
top_3_produits = ventes_par_produit.head(3).index.tolist()
print(top_3_produits)

final = merged[merged["product_category_name"].isin(top_3_produits)]

ventes_annee_produits = (
    final.groupby(["year", "product_category_name"])["price"]
    .sum()
    .reset_index()
)

ventes_annee_produits = ventes_annee_produits.pivot(index="year", columns="product_category_name", values="price")
ventes_annee_produits.plot(kind="bar", figsize=(12,6))
```

Ce diagramme à barres montre une croissance des ventes à partir de l'année 2017, en effet les ventes en 2016 sont quasiment négligeables pour les 3 catégories de produits et les ventes progressent encore plus en 2018. De plus, les ventes pour ces trois produits sont similaires pour l'année 2017, mais les produits de beauté et santé 'beleza_saude' se vendent mieux en 2018 tandis que les ventes des produits de literie,table et bain 'cama_mesa_banho' stagnent.
